<b>Get the data and store it in your pandas dataframe

```python
import pandas as pd
import numpy as np


df1=pd.read_csv('data/ratings_small.csv')
df2=pd.read_csv('data/links_small.csv')
df3=pd.read_csv('data/movies_metadata.csv',encoding='latin-1')

```

## 1. Check what is inside the data

```python
print(df3.head(3))
```

<b>Rename the id in df3 into movieId because we are going to use this key to join all the dataframe together

```python
df3.rename(columns = {'id':'movieId'},inplace=True)
```

<b> merge all dataframe together, 
    <br>since after observing the data we know that df3 already have id, which means movieId, then we just need to merge df1 and df3

<b>Cast the key=movieId into the same type (whether int or object)

```python
df1['movieId']=df1['movieId'].astype('str')
df3['movieId']=df3['movieId'].astype('str')
```

```python
df=pd.merge(df1,df3,on='movieId')
```

```python
df1.merge(df3, on='movieId')
```

<b> Count how many missing values

```python
df[df['runtime'].isnull()]
```

<b> Deal with missing value

```python
m_runtime=np.mean(df['runtime'].astype(float))
df['runtime'].fillna(m_runtime,inplace=True)
```
